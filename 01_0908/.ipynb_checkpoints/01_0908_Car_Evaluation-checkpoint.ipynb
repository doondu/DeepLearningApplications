```python
import numpy as np

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
column_names = ["buying", "maint", "doors", "persons", "lug_boot", "safety", "class"]
df_data = pd.read_csv("Car_Dataset/car_evaluation.csv", header=None, names=column_names)
df_data.shape
```

```python
df_data
```

```python
df_data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

features_to_encode = column_names

for feature in features_to_encode:
    df_data[feature] = label.fit_transform(df_data[feature])

df_data
```

```python
x = df_data
x = df_data.drop(columns=["class"], axis=1)
x
```

```python
y = df_data['class']
y
```

```python
# train & test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# SVM

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

clf_svm = SVC(random_state=0)
clf_svm.fit(x_train, y_train)

pred_svm = clf_svm.predict(x_test)

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

```python
# LR

from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(x_train, y_train)

pred_lr = clf_lr.predict(x_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
#DT

from sklearn.tree import DecisionTreeClassifier

clf_df = DecisionTreeClassifier(random_state=0)
clf_df.fit(x_train, y_train)

pred_df = clf_df.predict(x_test)

print ("\n--- Decision Tree Classifier ---")
print(accuracy_score(y_test, pred_df))
print(confusion_matrix(y_test, pred_df))
```

```python
# RT

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

print ("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(x_train, y_train)
pred = rf_clf.predict(x_test)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```
