```python
import numpy as np

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
column_names = ["buying", "maint", "doors", "persons", "lug_boot", "safety", "class"]
df_data = pd.read_csv("../01_0908/Car_Dataset/car_evaluation.csv", header=None, names=column_names)
df_data.shape
```

```python
df_data
```

```python
df_data.isnull().sum() #결측값 확인, 없음
```

```python
#data 모두 숫자로 변경, 딥러닝은 Class뿐 아니라 모든 데이터 숫자형 변경 필요
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

features_to_encode = column_names

for feature in features_to_encode:
    df_data[feature] = label.fit_transform(df_data[feature])

df_data
```

```python
#EDA
#unbalancing이 심한 데이터
sns.countplot(data = df_data, x="class")
plt.ylabel("Count")
plt.title("Label")
```

```python
df_corr = df_data.corr()

plt.figure(figsize=(10, 10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()
```

```python
#Splitting the data into training and test test
X = df_data.iloc[:,0:5].values
Y = df_data.iloc[:,6].values
```

```python
#one hot encoding (regression은 output이 1개이므로 필요 없음)
from tensorflow.keras.utils import to_categorical

y = to_categorical(Y)
```

```python
X.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD,Adam


model = Sequential()

model.add(Dense(5,input_shape=(5,),activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(4,activation='sigmoid'))

model.compile(Adam(learning_rate=0.02),'categorical_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
model_history=model.fit(x=X_train, y=y_train, epochs=100, batch_size=32,validation_data= (X_test, y_test))
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

loss =model_history.history['loss']
val_loss =model_history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
acc =model_history.history['accuracy']
val_acc =model_history.history['val_accuracy']
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
#Accuracy of the predicted values
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class))
print(confusion_matrix(y_test_class,y_pred_class))
```
