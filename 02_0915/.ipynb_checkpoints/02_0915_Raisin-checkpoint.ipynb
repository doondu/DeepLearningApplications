```python
import numpy as np

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df_data = pd.read_excel("Raisin_Dataset/Raisin_Dataset.xlsx")
df_data.shape
```

```python
df_data
```

```python
df_data.isnull().sum() #결측값 확인, 없음
```

```python
#Class 숫자로 변경
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

df_data['Class'] = label.fit_transform(df_data['Class'])

df_data
```

```python
#EDA
sns.countplot(data = df_data, x="Class")
plt.ylabel("Count")
plt.title("Label")
```

```python
df_corr = df_data.corr()

plt.figure(figsize=(10, 10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()
```

```python
#Splitting the data into training and test test
X = df_data.iloc[:,0:6].values
Y = df_data.iloc[:,7].values
Y
```

```python
#one-hot encoding, Numpy + one-hot encoding
from tensorflow.keras.utils import to_categorical
num_classes = 2
y = to_categorical(Y, num_classes = num_classes)
y
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

# 나눈 데이터의 크기 확인
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## FCN, Dense Layer

```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, SimpleRNN, LSTM
from keras.optimizers import SGD,Adam
from keras.callbacks import ModelCheckpoint
from sklearn.metrics import f1_score

model0 = Sequential()

model0.add(Dense(6,input_shape=(6,),activation='relu'))
model0.add(Dense(6,activation='relu'))
model0.add(Dense(6,activation='relu'))
model0.add(Dense(2,activation='sigmoid'))

model0.compile(Adam(learning_rate=0.02),'binary_crossentropy',metrics=['accuracy'])

model0.summary()
```

```python
# Best F1 Score, callback function
model0_history=model0.fit(x=X_train, y=y_train, epochs=100, batch_size=32,validation_data= (X_test, y_test))
y_pred = model0.predict(X_test)
```

```python
y_pred
```

```python
#형태 변환
y_test_class = np.argmax(y_test, axis=1)
y_pred_class = np.argmax(y_pred, axis=1)

print("실제 클래스:", y_test_class)
print("예측 클래스:", y_pred_class)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

loss =model0_history.history['loss']
val_loss =model0_history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
acc =model0_history.history['accuracy']
val_acc =model0_history.history['val_accuracy']
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
#Accuracy of the predicted values
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class))
print(confusion_matrix(y_test_class,y_pred_class))
```
